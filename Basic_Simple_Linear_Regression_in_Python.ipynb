```python
# Import libraries
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
# Get dataset and display
# for google colab you need to upload dataset as .csv file then put path of file here
# this dataset is relationship between TV Ads Budget and Sales
# we wnat to know if we paid for TV advertisements, how can sales be predicted?
df = pd.read_csv('/content/Ads Budget and Sales.csv')
print(df.head())
# Describe data
print(df.describe())
# Count data
print(df.count())
```

```python
# plot a scatter
# We choose only 2 columns , TV Ad Budget and Sales
sns.scatterplot(data=df, x="TV Ad Budget", y="Sales")
```

```python
# Splitting variables
X = df.iloc[:, 1:2]  # is the independent variable (TV Ad Budget)
y = df.iloc[:, 4:5]  # dependent on experience TV Ad Budget this y is what we have to predict
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = reg.predict(X_test)     # predicted value of y_test
y_pred_train = reg.predict(X_train)   # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'darksalmon')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('TV Ad Budget vs Sales (Training Set)')
plt.xlabel('TV Ad Budget')
plt.ylabel('Sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'TVA/Sal', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Prediction on test set
plt.scatter(X_test, y_test, color = 'darksalmon')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('TV Ad Budget vs Sales (Test Set)')
plt.xlabel('TV Ad Budget')
plt.ylabel('Sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'TVA/Sal', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Our input is 200 usd paid for TV Ads and predict how can sales
budget = 200
predict = reg.predict([[budget]])
print(f'Your budget for TV advertisements is {budget} and our predict for sales is {math.trunc(predict[0][0])}')

```
