```python
# performing linear algebra
import numpy as np

# data processing
import pandas as pd

# visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# Splitting data to training and testing
from sklearn.model_selection import train_test_split

# Knn classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("data.csv")

df.head()
```

```python
df.info()
```

```python
# dropping columns – ‘id’ and ‘Unnamed: 32’ as they have no role in prediction

df = df.drop(['Unnamed: 32', 'id'], axis = 1)
print(df.shape)
```

```python
# Converting the diagnosis value of M and B to a numerical value where M (Malignant) = 1 and B (Benign) = 0
def diagnosis_value(diagnosis):
	if diagnosis == 'M':
		return 1
	else:
		return 0

df['diagnosis'] = df['diagnosis'].apply(diagnosis_value)
df.head()
```

```python
# plot graph
sns.lmplot(x = 'radius_mean', y = 'texture_mean', hue = 'diagnosis', data = df)


```

```python
# X = all rows , all columns except ['diagnosis']
# y = all rows with only column ['diagnosis']
X = np.array(df.iloc[:, 1:])
y = np.array(df['diagnosis'])
```

```python
# split data test and train , test 33% with random = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# make knn model

knn = KNeighborsClassifier(n_neighbors = 13)
knn.fit(X_train, y_train)
```

```python
# predict
knn.score(X_test, y_test)
```
